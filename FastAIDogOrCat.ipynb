```python
# Importing all the necessary libraries:
import fastbook
from fastai.vision.all import *
import ipywidgets as widgets
```

```python
#
fastbook.setup_book()
path = untar_data(URLs.PETS)
```

```python
# Determining the labels:
def is_cat(x):
  return x[0].isupper()
```

```python
# Rapping around several data loaders for the problems of computer vision:
dls = ImageDataLoaders.from_name_func(
    path,
    get_image_files(path),
    valid_pct = 0.2,
    seed = 42,
    label_func = is_cat,
    item_tfms = Resize(224)
)
```

```python
# Training the model and making the predictions:
learn = cnn_learner(dls, resnet34, metrics = error_rate)

learn.fine_tune(1)

uploader = widgets.FileUpload()
uploader
def prediction():
  image = PILImage.create(uploader.data[0])
  image.show()

  # Making prediction
  is_cat,_,probs = learn.predict(image)

  print(f'Image is of a cat: {is_cat}.')
  print(f'Probability image is a cat: {probs[1].item:.6f}')
  prediction()
```
